In the dataset provided, each row represents a resume. The 'race' column has two values, 'b' and 'w', indicating black-sounding and white-sounding. The column 'call' has two values, 1 and 0, indicating whether the resume received a call from employers or not.

Note that the 'b' and 'w' values in race are assigned randomly to the resumes when presented to the employer.

1. Given that our data is categorical variables, a binomial test is be appropriate to determine if there is a statistically significant difference between the callback rates black and white candidates. CLT does apply because the 'black' and 'white' sound names were randomly assigned before presenting to the employer, in addition to the sample size less than 10% of the total universe of resumes submitted. 

2. The null hypothesis is that there is no difference between the black and white names, while the alternative hypothesis is that there is a difference between the two name types.

3. Margin of error is 1%, the confidence interval at a 99% confidence level is 5-8%, and the probability is 2e-6 %

4. There is a significant statistical difference of call back rates between black and white resume names. The probability of the observed difference of 4% between the two populations had 2e-6 % of occuring by chance. That is incredibly low! 

5. Our analysis does not mean that race/name is the most important factor in callback success, but perhaps one of the many features that indicates a difference between the two populations. Just because we observe a correlation between the variables does not imply causation. There could be a confounding variable such as education or age that may provide a more rational explanation for discrepancies between skin color. 



```python
import pandas as pd
import numpy as np
from scipy import stats
data = pd.io.stata.read_stata(r'D:\Data Science\Springboard\racial_disc\data\us_job_market_discrimination.dta')
```

```python
# number of callbacks for black-sounding names
w_callback = sum(data[data.race=='w'].call)
w_prob = w_callback/2435
print('baseline probability of being called back for an interview is: %.2f' % w_prob)

b_callback = sum(data[data.race=='b'].call)
b_prob = b_callback/2435
print('probability of being called back and being black is %.2f' % b_prob)

p = w_prob
q = 1-p
n = 2435

standard_error = np.sqrt(p*q/n)
print('standard error: %.2f' % standard_error)

binomial_test = stats.binom_test(b_callback, n = 2435, p = w_prob)
print('probability: %s' % binomial_test)

confidence_interval = standard_error*3 #99% confidence interval 
print('confidence interval: %.2f - %.2f' % (b_prob - confidence_interval, b_prob + confidence_interval))
```
